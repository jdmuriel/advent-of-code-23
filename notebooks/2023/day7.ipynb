```python
#lower is better
types={'five':0,'four':1,'full':2,'three':3,'twopair':4,'onepair':5,'high':6}
cards={'A':10,'K':11,'Q':12,'J':13,'T':14,'9':15,'8':16,'7':17,'6':18,'5':19,'4':20,'3':21,'2':22,'1':23}
cards2={'A':10,'K':11,'Q':12,'T':14,'9':15,'8':16,'7':17,'6':18,'5':19,'4':20,'3':21,'2':22,'1':23,'J':24}

def gethist(hand):
    hist = {}
    for card in hand:
        hist[card] = hist.get(card,0)+1
    hist = [(card,count) for card,count in hist.items()]
    hist = sorted(hist,key=lambda pair: pair[1], reverse=True)
    return hist

def gethist2(hand):
    hist = {}
    for card in hand:
        hist[card] = hist.get(card,0)+1
    hist = [[card,count] for card,count in hist.items()]
    hist = sorted(hist,key=lambda pair: pair[1], reverse=True)
    # apply jokers
    hist2 = []
    maxcard = hist[0][0]
    if maxcard == 'J' and len(hist)==1:
        # poker of J becomes poker of As
        hist2.append(['A',5])
    elif maxcard == 'J':
        # remove Js and add to second card
        hist2.append([hist[1][0],hist[0][1] + hist[1][1]])
        for card,count in hist[2:]:
            hist2.append([card,count])
    else:
        for card,count in hist:
            if card == 'J':
                #convert J to maxcard
                hist2[0][1] = hist2[0][1] + count
            else:
                hist2.append([card,count]) 
    if sum([count for x,count in hist2]) != 5:
        return "error"
    return hist2


def gettype(hand, gethistfn=gethist):
    hist : dict = gethistfn(hand)
    #print(hist)
    if hist[0][1] == 5:
        return 'five'
    elif hist[0][1] == 4:
        return 'four'
    elif hist[0][1] == 3 and hist[1][1] == 2:
        return 'full'
    elif hist[0][1] == 3:
        return 'three'
    elif hist[0][1] == 2 and hist[1][1] == 2:
        return 'twopair'
    elif hist[0][1] == 2:
        return 'onepair'
    else:
        return 'high'

#def compare(hand1, hand2):
x = gethist("AKK01")
print(x)
print(gettype("AK0KK1"))
```

```python
def hand_str(hand:str, cards):
    result = ''
    for card in hand:
        result += str(cards[card])
    return result

# comparable value: type_value+"-"+hand_str
def key(hand):
    return str(types[gettype(hand)]) + '-' + hand_str(hand, cards)

def key2(hand):
    return str(types[gettype(hand, gethist2)]) + '-' + hand_str(hand, cards2)
```
