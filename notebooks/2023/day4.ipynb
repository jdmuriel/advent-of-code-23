```python
import re

result = 0
cards = {x: 1 for x in range(0,len(lines))}

for i in range(0,len(lines)):
    line=lines[i]
    card, numbers = line.replace('\n','').split(': ')
    winning,playing = numbers.split('| ')
    winning_set:set = {int(x) for x in re.split(r'\s+',winning.strip())}
    playing = [int(x) for x in re.split(r'\s+',playing.strip())]
    wins = [x for x in playing if x in winning_set]
    if wins:
        value = pow(2,len(wins)-1)
        print (f"Wins card {i+1}: {len(wins)}: {wins} with value {value}")
        result += value
        for nextcard in range(i+1,i+len(wins)+1):
            # Subimos el nº de cartas
            print(f"Agregamos {cards[i]} cartas a la fase {nextcard+1}")
            cards[nextcard]=cards[nextcard]+cards[i]
        
print (cards)
print (sum(cards.values()))
print (result)
```
