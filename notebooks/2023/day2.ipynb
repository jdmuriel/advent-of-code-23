```python
import os
file_path = os.path.join(os.curdir, "day2.txt")
with open(file_path) as file:
    lines = file.readlines()

f"line count:{len(lines)}, first line:{lines[0]}"
```

```python
import re
def parseLine(line: str):
    # returns an int with number of the game, and a list of maps with the ball_count for each color

    match = re.match(r"Game (\d+): (.*)", line)
    game_number = int(match.group(1))
    game_str = match.group(2)

    extraction_list = []
    for extraction_str in game_str.split("; "):
        # extraction_str is something like "1 green, 3 red, 6 blue"
        extraction = {}
        for ball_str in extraction_str.split(", "):
            #ball_str is like "1 green"
            ball_count = int(ball_str.split(" ")[0])
            ball_color = ball_str.split(" ")[1]
            extraction[ball_color] = ball_count
        extraction_list.append(extraction)
    return game_number, extraction_list
```

```python
for line in lines:
    print(line)
    game_number, extraction_list = parseLine(line)
    print(f"{game_number} - {len(extraction_list)} - {extraction_list[0]}")
```

```python
def is_valid(extraction:dict):
    return extraction.get('red',0)<=12 and extraction.get('green',0)<=13 and extraction.get('blue',0)<=14

def is_valid_game(extraction_list:list):
    for extraction in extraction_list:
        if not is_valid(extraction):
            # print(f"game{game_number} with extraction {extraction} invalid in line {line}")
            return False
    return True
```

```python
def get_power(extraction_list:list):
    min_balls = {}
    for extraction in extraction_list:
        for color, count in extraction.items():
            if min_balls.get(color,0) < count:
                min_balls[color] = count
    power = min_balls.get('red',0)*min_balls.get('blue',0)*min_balls.get('green',0)
    print(f"game {game_number} sol {min_balls}, power {power} for line {line}")
    return power 
```

```python
# games is a list
# each game is a list of extractions
# each extraction is a map of "color" to int representing the number of balls of each color 
games = []
sum = 0
power_sum = 0
for line in lines:
    game_number, extraction_list = parseLine(line)
    # if is_valid_game(extraction_list):
    #     sum+=game_number
    power_sum += get_power(extraction_list)
print(power_sum)
```
