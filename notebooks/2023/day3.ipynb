```python
import re
def parseLine(line: str):
    # returns an int with number of the game, and a list of maps with the ball_count for each color

    match = re.match(r"Game (\d+): (.*)", line)
    game_number = int(match.group(1))
    game_str = match.group(2)

    extraction_list = []
    for extraction_str in game_str.split("; "):
        # extraction_str is something like "1 green, 3 red, 6 blue"
        extraction = {}
        for ball_str in extraction_str.split(", "):
            #ball_str is like "1 green"
            ball_count = int(ball_str.split(" ")[0])
            ball_color = ball_str.split(" ")[1]
            extraction[ball_color] = ball_count
        extraction_list.append(extraction)
    return game_number, extraction_list
```

```python
def is_valid(map, index, start, end):
    # print(f"index:{index} start:{start} end:{end}")
    charsToCheck = map[index-1][start-1:end+1] \
            + map[index][start-1] + map[index][end] \
            + map[index+1][start-1:end+1]
    for char in charsToCheck:
        if not char.isdigit() and char != '.':
            # print(f"Found {char} in {charsToCheck}")
            return True
    return False
```

```python
def add_gear_candidate(gears:dict, i,j,number):
    print(f"gear candidate at pos {i},{j} for number{number}")
    numbers=gears.get((i,j),[])
    numbers.append(number)
    gears[(i,j)] = numbers

def add_gears(map, index, start, end, gears, number):
    print(f"index:{index} start:{start} end:{end}")

    for j in range(start-1,end+1):
        if map[index-1][j] == '*':
            add_gear_candidate(gears,index-1,j,number)
    if map[index][start-1] == '*':
        add_gear_candidate(gears,index,start-1,number)
    if map[index][end] == '*':
        add_gear_candidate(gears,index,end,number)
    for j in range(start-1,end+1):
        if map[index+1][j] == '*':
            add_gear_candidate(gears,index+1,j,number)
```

```python
mysum=0
gears={}    # gears is a map with keys tuples(i,j) and keys an array of numbers
for index in range(1,len(map)):
    line=map[index]
    print(index, line)
    matches = re.finditer(r'\d+', line)
    for match in matches:
        if is_valid(map, index, match.start(), match.end()):
            mysum += int(match.group())
            add_gears(map, index, match.start(), match.end(), gears, int(match.group()))
    print(line)
print(f"sum:{mysum}")
print(gears)
```
